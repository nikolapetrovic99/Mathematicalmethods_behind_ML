### Plot

Osnovna naredba za linijski grafiki prikaz ili prikaz putem markera je [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). I ostale naredbe upotrebljene u narednim primerima [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) i [fill](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html) koriste se na slian nain i poseduju sline opcije podeavanja. Zbog toga emo prikazati samo naredbu `plot`. 

Najjednostavniji sintaksni oblik ove naredbe je
```python
matplotlib.pyplot.plot([x,] y [, opis formatiranja prikaza i/ili podataka])
```
Za naredbu `plot` osnovni ulazni podaci su koordinate taaka $(x_k,y_k),\ i=0,1,\dots,n.$ 
One se unose kao dve zasebne liste ili niza, vektor $x-$koordinata $x=\begin{bmatrix}x_0&x_1&\dots&x_n\end{bmatrix}^T,$ i vektor $y-$koordinata $y=\begin{bmatrix}y_0&y_1&\dots&y_n\end{bmatrix}^T.$ Kada je naveden samo jedan vektor podrazumeva se da je to vektor $y-$koordinata. U tom sluaju $x$ je indeksni skup, tj. $x=\begin{bmatrix}0&1&\dots&n-1\end{bmatrix}^T.$

Opisom `formatiranja prikaza` biramo nain kako e podaci biti grafiki predstavljeni: 
- prikaz taaka markerima odreene boje i oblika (prikaz isticanjem taaka),
- povezivanje taaka linijom odreene boje i oblika (linijski prikaz),
- kombinovani prikaz. 

Ukoliko se izostavi ovo opciono polje, podrazumeva se linijski prikaz podataka.


**Primer 1.** Najjednostavnije generisan grafiki prikaz linijski povezanih podataka dat je sledeom kodnom linijom. S obzirom da je naveden samo jedna lista vrednosti podrazumeva se da su to vrednosti $y$ koordinata tachaka.

````python
plt.plot([-1,1,3,-4,2,1]);  #kada ne pravimo sloene prikaze, Jupyter ima mogunost pojednostavljenog poziva crtanja
````

Kombinovanjem podataka za $x$-osu sa prethodnim $y$ vrednostima mogue je kreirati razliite oblike. 

````python
plt.plot([1,1,2,2.5,2,4],[-1,1,3,-4,2,1]);
````

Ukoliko elimo da grafik sadri neto vie podataka, potrebno je razumevanje pojedinanih delova objekta `figure`. Grafiki prikaz moe da sadri jednu ili vie zasebnih slika - odeljaka. Jedan odeljak **subplot** ili **axes** sadri naslov **title**, granice slike **spines** (top, bottom, left, right), i dve ose: $x$ i $y$ osu **axis**. 

<div class="alert alert-block alert-info">
<b>Napomena:</b> Obratite panju na razliku u imenima <b>axes</b> i <b>axis</b>. </div>

Koordinatne ose **axis** slue da prikau raspon u vrednostima grafika koji se prikazuje. Ose mogu da nose svoje ime **label**. Osim toga, i odeljak, tj. **axes** mogu da nose ime kroz opciju `title`. 

Za lake itanje vrednosti sa grafika koristi se mrea horizontalnih i vertikalnih linija vezane za istaknute vrednosti na $x$ i $y$ osi. Ova mrea aktivira se opcijom `grid`. Istaknuti podeoci na $x$ i $y$ osi nose ime `ticks`.

````python
plt.plot([-1,1,3,-4,2,1])       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.grid()
````

`xlabel(string)` postavlja svoj string argument za ime $x-$ose na grafiku. Slino, `ylabel(string)` slui za imenovanje $y-$ose na grafiku ili podgrafiku.

Kako bismo postigli uobiajen prikaz grafika sa istaknutim pozicijama $x$ i $y$ ose na samom grafiku moemo da koristimo `axhline()` i `axvline()` opcije. `axhline()` crta horizontalnu liniju $y=0.$ Osim toga `axhline(y=a)` iscrtava horizontalnu liniju $y=a$, gde je $a$ proizvoljna numerika veliina. Opcije boja definiu se kroz argument `color` ija se vrednost postavlja na string kojim se navodi ime konkretne boje na engleskom jeziku. Po istom principu funkcionie i naredba za vertikalne linije  `axvline()`.

U primeru je prikazana i upotreba opcionog argument `zorder` kojim se definie redosled prikaza na jednoj slici. Moe se navoditi u okviru razliitih naredbi grafikog prikaza. Podrazumeva se da se sa svakim novim grafikim elementom koji se doda na sliku on postavlja ispred prethodnih. Zbog toga se koristi `zorder` kao dodatna opcija za definisanje slojeva na slici. Ovaj argument `zorder` moe da uzme neku celobrojnu vrednost. to je vea vrednost ovog argumenta za neki grafiki element to je on na viem sloju prikaza. Matplotlib automatski postavlja `zorder=0` za elemente slike. Zbog toga postavljanje `zorder=-1` obezbeuje da se odgovarajui element na slici prikazuje iza svih ostalih, barem onih koji imaju vei ovaj argument.

````python
plt.plot([-1,1,3,-4,2,1])       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.axhline(color='black', zorder=-1)
plt.axhline(y=-3, color='blue', zorder=1)
plt.axvline(color='red', zorder=2);
plt.axvline(x=2,color='green', zorder=-1);
````

U okviru jednog grafikog odeljka (subplot, axes) mogu da se prikau grafiki izlazi vie naredbi. U situaciji kada jedna slika ili podslika sadri kombinovano vie razliitih grafika korisno je da postoji [legenda](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) u kojoj se objanjava izvor odgovarajueg skupa podataka. Imena odgovarajuih grafikih podataka za legendu mogu da se definiu ve kroz same naredbe crtanja definisanjem opcionog parametra `label`.

````python
plt.plot([-1,1,3,-4,2,1], label="prva serija", zorder=1)    #opcija label imenuje podatke za legendu
plt.scatter(range(6),[0,0,1,-1,3,-3],label="druga serija")   #takasti prikaz podatka  
plt.fill(range(4),[-1,-3,-2,0], label="poligon")             #prikaz popunjavanjem
plt.legend(loc='upper left', title='legenda');
````

Pozivom naredbe `legend()` lista prikazanaih podataka sa imenom i oznakom na grafiku postake deo grafikog prikaza. Svaki skup $x-y$ podataka imenuje se na osnovu stringa koji je definisan kljunom rei `label` u samom pozivu funkcije
`plot`. Pozicija legende moe da se korisniki definie argumentom `loc`. Rezervisana re `title` moe da se koristi za definisanje imena legende. O svemu tome moe da se konsultuje [dokumentacioni centar](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

## Ureivanje sadraja grafikog prozora

**Primer 2.** Upoznaemo jo neke karakteristike funkcije `plot` kroz kratak, jednostavan primer. Pretpostavimo da su dati podaci o dnevnoj prodaji softvera: Digital Lab, Mobile Transfer, Advanced System Care i Backup Pro 14. Podaci se odnose na jednu radnu nedelju, tj. period ponedeljak-petak. Naredne liste nose te podatke. 

````python
Dani=range(5)                         #naredba range proizvodi liste na isti nain kao to arange kreira NumPy nizove
Digital_Lab=[12,7,9,3,-10]
Mobile_Transfer=[3,2,5,0,-5]
Advanced_System_Care=[1,5,0,3,-2]
Backup_pro14=[1,0,2,0,0]
````

<div class="alert alert-block alert-info">
<b>Napomena:</b> Umesto lista mogli smo da koristimo i NumPy nizove, ukoliko bismo taj modul prethodno uitali.  </div>

U Jupyter radnim sveskama sve informacije koje elimo da prikaemo na jednom grafiku, naredbe za izvrenje njihovih prikaza moraju da budu deo jedne iste kodne elije. Postavljanjem i brisanjem oznaka za komentare `#` u kodnoj eliji ovog primera moete lake da uoite ta svaka od naredbi ponaosob dodaje grafiku.

Svaka od osa podeljena je manjim i veim oznakama [ticks](https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html) (crticama) na dui jednakih duina. Duina crtica ukazuje da Matplotlib podrava dva nivoa podele osa - glavni (`major`, uobiajeno aktivan) i sporedni (`minor`, uobiajeno neaktivan). Aktivni podeoci koriste se za povlaenje mree linija `grid`. 

````python
slika, ax = plt.subplots(figsize=(8,6))          #definisanje imena objekta figure, imena odeljaka (axes) i njihovog broja-0
plt.title("Linijski prikaz prodaje i nabavke")
plt.plot(Digital_Lab, label='Digital Lab')         #opcija label imenuje podatke za legendu   
plt.xlabel("Dani u nedelji")
plt.ylabel("Prodaja po komadu")
plt.plot(Mobile_Transfer, label='Mobile Transfer')
plt.plot(Advanced_System_Care,label='Advanced System Care')
plt.plot(Backup_pro14,label='Backup Pro 14')
plt.xticks(Dani,["Ponedeljak","Utorak","Sreda","etvrtak","Petak"])
plt.legend(loc='lower left', shadow=True)
ax.yaxis.set_minor_locator(plt.MultipleLocator(1))
plt.grid(b=True, which='major', linestyle='-')
plt.grid(b=True, which='minor', linestyle='--')
````

**Primer 3.** Neka je data funkcija $f(x)=x^2\cos(x-2)$ na domenu $x\in[-3,6].$ Nacrtaemo linijski grafik ove funkcije na osnovu $100$ taaka vrednosti. Zbog velikog broja vrednosti, take sa $x-$ose ne unosimo runo, ve geberiemo nekom naredbom za automattsko generisanje nizova (ili lista).

````python
import numpy as np
````

````python
domen=np.linspace(-3,6,100)   #kreiranje x-podatka
fx=np.cos(domen-2)*(domen**2) #kreiranje y-podatka
````

Ovim smo pripremili podatke za crtanje grafika. Ukoliko primenimo naredbu **plot** bez navoenja $x-$vrednosti dobiemo grafik koji nije u potpunosti odgovarajui.

````python
plt.plot(fx)
plt.title("Grafik van domena funkcije");
````

Ukoliko uvrstimo i take domena u poziv naredbe dobiemo adekvatan raspored taaka sa grafika.

````python
plt.plot(domen,fx)
plt.title("Grafik na domenu funkcije")
plt.grid()
````

**Primer 4.** Na primeru grafika funkcije 
$$ f(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}},\qquad x\in[-5,5]$$
pokazaemo uticaj finoe gustine uzorkovanja kontinualne funkcije na izgled linijskog grafika.

````python
from ipywidgets import interactive

def f(n):
    x = np.linspace(-5, 5, n)
    y = np.exp(-x**2/2)/np.sqrt(2*np.pi)
    plt.title("Gausovo zvono")
    plt.plot(x, y,'-')
    plt.ylim(-0.3, 0.6)
    plt.grid()

zvono = interactive(f, n=(5,50, 5))
zvono
````

**Primer 5.** Iskoristiemo specijalizovane funkcije NumPy modula za rad sa polinomima za njihove grafike prikaze. Za poetak posmatraemo polinom $P_5(x)=3x^5-2.1x^4+1.12x^3-3.74,$ zadat svojim vektorom koeficijenata. Potraiemo njegov grafik na segmentu $[-1,2].$

````python
P5=np.array([3,-2.1,1.12,0,0,-3.74])
xk=np.linspace(-1,2)
fk=np.polyval(P5,xk)
````

````python
plt.plot(xk,fk)
plt.title("Grafik polinoma")
plt.axhline(color='black', zorder=-1)
plt.axvline(color='black', zorder=-1)
plt.grid();
````

**Primer 6.** Vektorski nain zadavanja koordinata taaka za Matplotlib modul pogoduje i prilikom crtanje grafika parametarski zadatih krivih. Vie o otome bie reeno u nekoj od narednih tema.

````python
def krug(n):
    t=np.linspace(0,2*np.pi,n)
    x_krug=6*np.cos(t)
    y_krug=6*np.sin(t)
    plt.plot(x_krug,y_krug,'k')
    plt.title("Krunica")
    plt.axis('equal')
    plt.xlim(-7,7)
    plt.ylim(-7,7)
    plt.grid()

interaktivna_slika = interactive(krug, n=(5,75, 5))
interaktivna_slika   
````

**Primer 7.** U narednom primeru kombinovaemo prethodno opisane prikaze i opcije. Primenjena je i biblioteka `random` kako bi se lake generisali podaci za prikaz.

````python
import numpy.random as rndm
````

````python
x = np.linspace(-10., 10., 200)
y = np.sin(x) * np.exp(-(x/5.0)**2)
yerror=rndm.uniform(-0.1,0.1,[20])

xdata= np.linspace(-10., 10., 20)
ydata = np.sin(xdata) * np.exp(-(xdata/5.0)**2)+yerror

plt.figure(1, figsize=(6, 4))
plt.plot(x, y, 'g-', label='teorija') 
plt.plot(xdata, ydata, 'bo', label="podaci") 
plt.xlabel('x')
plt.ylabel('transverzalno izmetanje')
plt.legend(loc='upper right', title='legenda')
plt.axhline(color='gray', zorder=-1)
plt.axvline(color='gray', zorder=-1);
````

Podaci prikazani na prethodnoj slici imitiraju situaciju eksperimentalnih podataka sa pokuajem njihovog modeliranja, tj. pronalaenja funkcije koja e ih pratiti na neki blizak nain. Takve funkcije koristimo za priblinu dopunu informacija o praenom fenomenu. Na gornjoj slici takasto su prikazani mereni podaci - eksperimentalni podaci, dok je zelenom linijom dat model koji dalje vrednosti merene veliine imitira i predvia. U takvim situacijama veoma esto se praktikuje ukljuivanje podataka o odstupanjima vrednosti eksperimenta i njegovog modela. To nam pomae, izmeu ostalog, u proceni valjanosti dobijenog modela. Modul Matplotlib poseduje funkciju `errorbar` koja takaste podatke prikazuje sa horizontalnim i/ili vertikalnim linijama koje slue za opis greke. Narednim primerom prikazana je upotreba ove funkcije.

````python
plt.figure(1, figsize=(6, 4))
plt.plot(x, y, '-C0', label="teorija")
plt.errorbar(xdata, ydata, fmt='oC1', label="podaci",
xerr=0.75, yerr=yerror, ecolor='black')
plt.xlabel('x')
plt.ylabel('transverzalno izmetanje')
plt.legend(loc='upper right');
````

**Primer 8.** Prikazaemo grafik funkcije $\tan x$ za vrednosti argumenta $x\in[0,10],$ jasno je radi se o uglovima u radijanima. To moemo da pokuamo da obavimo bez mnogo ureivanja na ve prikazan nain, jednim pozivom naredbe `plot` bez dodatnih argumenata, kao to je uraeno u narednoj kodnoj eliji.

````python
x = np.arange(0.01, 10., 0.04)
ytan = np.tan(x)

plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytan);
````

Rezultujui grafik ne izgleda ba kako smo oekivali. Problem nastaje zbog toga to funkcija $\tan x$ poseduje vertikalne asimptote u takama $x\in\{\pm\pi/2,\pm3\pi/2,\dots\}.$ Drugim reima veliki rasponi vrednosti se registruju i prikazuju kao iljci na grafiku. Takoe neujednaenost grafika ukazuje na prekoraenje u predstavivim vrednostima i nestabilnost u izraunavanju. Naravno,takav grafik je od slabe koristi. Reenje nalazimo u ogranienju skupa vrednosti $y$ koordinata koje se prikazuju na grafiku. Tome slui Matplotlib funkcija `ylim` kao to je demonstrirano u narednoj eliji.

````python
plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytan)
plt.ylim(-8, 8)
plt.axhline(color="gray", zorder=-1);
````

Sa gornjeg grafika moemo i da iskljuimo vertikalne linije koje se pojavljuju u blizini taaka $x\in\{\pi/2,3\pi/2,5\pi/2\}.$ Tome slui naredba za [maskiranje](https://numpy.org/doc/stable/reference/generated/numpy.ma.masked_where.html) NumPy nizova. Primer upotrebe dat je u narednoj eliji. Prvi argument opisuje uslov za maskiranje elemenata nekog niza, sam niz se navodi kao drugi argument.

````python
ytanM = np.ma.masked_where(np.abs(ytan) > 20., ytan)

plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytanM)
plt.ylim(-8, 8) 
plt.axhline(color="gray", zorder=-1);
````

**Primer 10.** Kreiraemo nepravilnu tabelu grafikih elemenata. Nepravilnost se ogleda u nejednakom broju slika unutar vrsta tabela. Najpre kreiramo takaste podatke za grafiki prikaz. Funkcije koje e biti prikazane su $\quad s(x)=\begin{cases}1,&x=0,\\ \dfrac{\sin x}{x},&x\neq0,\end{cases}\quad$ i $\quad exp(x)=e^x.$

````python
#definisanje funkcije s(x)
def s(x):
    a = np.where(x == 0., 1., np.sin(x)/x)
    return a

#podaci za funkciju s
xs = np.linspace(-15., 15., 150)
ys = s(xs)

#podaci za eksponencijalnu funkciju
xe = np.arange(0., 10., 0.1)
ye = np.exp(xe)
````

````python
# kreiranje grafikog prozora izabrane veliine
slika = plt.figure(figsize=(9, 7))
plt.suptitle("Kombinovanje grafika 2+1")
slika.subplots_adjust(hspace=0.3, wspace = 0.3) #razmak izmeu delova prozora

# subplot: klasian prikaz eksponencijalne funkcije
plt.subplot(2, 2, 1)
plt.plot(xe, ye)
plt.title(r'$e^x$')

# subplot: y-log prikaz eksponencijalne funkcije
plt.subplot(2, 2, 2)
plt.plot(xe, ye)
plt.yscale("log")
plt.title(r'$e^x$ sa $\log$ razmerom na $y$ osi')

# subplot: prikaz s(x) funkcije
plt.subplot(2, 1, 2)
plt.plot(xs, ys)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.title(r's(x)=(\sin x)/x');

#slika.tight_layout()  #nain razmaka izmeu podslika
````

## Histogram

Za narednu temu bie koristan jedan poseban nain predstavljanja numerikih podataka - histogram. On predstavlja grafiku verziju tabele u kojoj se uvaju informacije o frekvenciji pojavljivanja vrednosti iz unapred izabranih intervala. Histogram je nain prikazivanja podataka rasporeenih u odreene kategorije (klase) ili grupe. Pojasniemo ovo na primeru.

**Primer 11.** Pretpostavimo da je dat vektor od 50 celobrojnih vrednosti u opsegu $[0,20].$ elimo da saznamo frekvenciju pojavljivanja vrednosti iz pojedinanih intervala $[0,5),[5,10),[10,15),[15,20].$ Takvom analizom moemo da utvrdimo da li su neki opsezi vrednosti zastupljeniji ili ne. 

Horizontalnim linijama na grafiku prikazana je podela vrednosti po grupama. 

````python
primer9, ax = plt.subplots()  
indeksi=np.arange(50)
vrednosti=np.random.randint(21,size=(50,))
plt.scatter(indeksi,vrednosti)
ax.yaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='y', linestyle='-')
````

Prebrojavanjem elemenata u svakoj od grupa dobijamo podatke o zastupljenosti vrednosti iz svake od grupa. Grafiki se to pokazuje visinom stubaca iznad odgovarajueg intervala.

````python
histogram, ax = plt.subplots()  
plt.hist(vrednosti, 4)
plt.title("Histogram")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='both', linestyle='-')
````

Ukoliko elimo da histagram pokazuje relativne vrednosti ponavljanja potrebno je postaviti opciju **density=True**.

````python
normalizovan_histogram, ax = plt.subplots()  
plt.hist(vrednosti, 4, density=True)
plt.title("Histogram")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='both', linestyle='-')
````
