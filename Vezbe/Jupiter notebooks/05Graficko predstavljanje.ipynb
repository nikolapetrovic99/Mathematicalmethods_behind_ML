## Crtanje 2D grafika

Naredbom `figure()` kreira se prazan prozor za grafički prikaz. Višestruki poziv funkcije  `figure()` otvara više prozora za grafički prikaz pa mogu da nose brojeve kao svoje labele kako bi se razlikovali. Tako npr. `figure(1)` otvara jedan prozor imena 1, `figure(2)` drugi i `figure(3)` još jedan. ### Plot

Osnovna naredba za linijski grafički prikaz ili prikaz putem markera je [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). I ostale naredbe upotrebljene u narednim primerima [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) i [fill](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html) koriste se na sličan način i poseduju slične opcije podešavanja. Zbog toga ćemo prikazati samo naredbu `plot`. 

Najjednostavniji sintaksni oblik ove naredbe je
```python
matplotlib.pyplot.plot([x,] y [, opis formatiranja prikaza i/ili podataka])
```
Za naredbu `plot` osnovni ulazni podaci su koordinate tačaka $(x_k,y_k),\ i=0,1,\dots,n.$ 
One se unose kao dve zasebne liste ili niza, vektor $x-$koordinata $x=\begin{bmatrix}x_0&x_1&\dots&x_n\end{bmatrix}^T,$ i vektor $y-$koordinata $y=\begin{bmatrix}y_0&y_1&\dots&y_n\end{bmatrix}^T.$ Kada je naveden samo jedan vektor podrazumeva se da je to vektor $y-$koordinata. U tom slučaju $x$ je indeksni skup, tj. $x=\begin{bmatrix}0&1&\dots&n-1\end{bmatrix}^T.$

Opisom `formatiranja prikaza` biramo način kako će podaci biti grafički predstavljeni: 
- prikaz tačaka markerima određene boje i oblika (prikaz isticanjem tačaka),
- povezivanje tačaka linijom određene boje i oblika (linijski prikaz),
- kombinovani prikaz. 

Ukoliko se izostavi ovo opciono polje, podrazumeva se linijski prikaz podataka.


**Primer 1.** Najjednostavnije generisan grafički prikaz linijski povezanih podataka dat je sledećom kodnom linijom. S obzirom da je naveden samo jedna lista vrednosti podrazumeva se da su to vrednosti $y$ koordinata tachaka.

````python
plt.plot([-1,1,3,-4,2,1]);  #kada ne pravimo složene prikaze, Jupyter ima mogućnost pojednostavljenog poziva crtanja
````

Kombinovanjem podataka za $x$-osu sa prethodnim $y$ vrednostima moguće je kreirati različite oblike. 

````python
plt.plot([1,1,2,2.5,2,4],[-1,1,3,-4,2,1]);
````

Ukoliko želimo da grafik sadrži nešto više podataka, potrebno je razumevanje pojedinačnih delova objekta `figure`. Grafički prikaz može da sadrži jednu ili više zasebnih slika - odeljaka. Jedan odeljak **subplot** ili **axes** sadrži naslov **title**, granice slike **spines** (top, bottom, left, right), i dve ose: $x$ i $y$ osu **axis**. 

<div class="alert alert-block alert-info">
<b>Napomena:</b> Obratite pažnju na razliku u imenima <b>axes</b> i <b>axis</b>. </div>

Koordinatne ose **axis** služe da prikažu raspon u vrednostima grafika koji se prikazuje. Ose mogu da nose svoje ime **label**. Osim toga, i odeljak, tj. **axes** mogu da nose ime kroz opciju `title`. 

Za lakše čitanje vrednosti sa grafika koristi se mreža horizontalnih i vertikalnih linija vezane za istaknute vrednosti na $x$ i $y$ osi. Ova mreža aktivira se opcijom `grid`. Istaknuti podeoci na $x$ i $y$ osi nose ime `ticks`.

````python
plt.plot([-1,1,3,-4,2,1])       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.grid()
````

`xlabel(string)` postavlja svoj string argument za ime $x-$ose na grafiku. Slično, `ylabel(string)` služi za imenovanje $y-$ose na grafiku ili podgrafiku.

Kako bismo postigli uobičajen prikaz grafika sa istaknutim pozicijama $x$ i $y$ ose na samom grafiku možemo da koristimo `axhline()` i `axvline()` opcije. `axhline()` crta horizontalnu liniju $y=0.$ Osim toga `axhline(y=a)` iscrtava horizontalnu liniju $y=a$, gde je $a$ proizvoljna numerička veličina. Opcije boja definišu se kroz argument `color` čija se vrednost postavlja na string kojim se navodi ime konkretne boje na engleskom jeziku. Po istom principu funkcioniše i naredba za vertikalne linije  `axvline()`.

U primeru je prikazana i upotreba opcionog argument `zorder` kojim se definiše redosled prikaza na jednoj slici. Može se navoditi u okviru različitih naredbi grafičkog prikaza. Podrazumeva se da se sa svakim novim grafičkim elementom koji se doda na sliku on postavlja ispred prethodnih. Zbog toga se koristi `zorder` kao dodatna opcija za definisanje slojeva na slici. Ovaj argument `zorder` može da uzme neku celobrojnu vrednost. Što je veća vrednost ovog argumenta za neki grafički element to je on na višem sloju prikaza. Matplotlib automatski postavlja `zorder=0` za elemente slike. Zbog toga postavljanje `zorder=-1` obezbeđuje da se odgovarajući element na slici prikazuje iza svih ostalih, barem onih koji imaju veći ovaj argument.

````python
plt.plot([-1,1,3,-4,2,1])       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.axhline(color='black', zorder=-1)
plt.axhline(y=-3, color='blue', zorder=1)
plt.axvline(color='red', zorder=2);
plt.axvline(x=2,color='green', zorder=-1);
````

U okviru jednog grafičkog odeljka (subplot, axes) mogu da se prikažu grafički izlazi više naredbi. U situaciji kada jedna slika ili podslika sadrži kombinovano više različitih grafika korisno je da postoji [legenda](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) u kojoj se objašnjava izvor odgovarajućeg skupa podataka. Imena odgovarajućih grafičkih podataka za legendu mogu da se definišu već kroz same naredbe crtanja definisanjem opcionog parametra `label`.

````python
plt.plot([-1,1,3,-4,2,1], label="prva serija", zorder=1)    #opcija label imenuje podatke za legendu
plt.scatter(range(6),[0,0,1,-1,3,-3],label="druga serija")   #tačkasti prikaz podatka  
plt.fill(range(4),[-1,-3,-2,0], label="poligon")             #prikaz popunjavanjem
plt.legend(loc='upper left', title='legenda');
````

Pozivom naredbe `legend()` lista prikazanaih podataka sa imenom i oznakom na grafiku postake deo grafičkog prikaza. Svaki skup $x-y$ podataka imenuje se na osnovu stringa koji je definisan ključnom reči `label` u samom pozivu funkcije
`plot`. Pozicija legende može da se korisnički definiše argumentom `loc`. Rezervisana reč `title` može da se koristi za definisanje imena legende. O svemu tome može da se konsultuje [dokumentacioni centar](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

## Uređivanje sadržaja grafičkog prozora

**Primer 2.** Upoznaćemo još neke karakteristike funkcije `plot` kroz kratak, jednostavan primer. Pretpostavimo da su dati podaci o dnevnoj prodaji softvera: Digital Lab, Mobile Transfer, Advanced System Care i Backup Pro 14. Podaci se odnose na jednu radnu nedelju, tj. period ponedeljak-petak. Naredne liste nose te podatke. 

````python
Dani=range(5)                         #naredba range proizvodi liste na isti način kao što arange kreira NumPy nizove
Digital_Lab=[12,7,9,3,-10]
Mobile_Transfer=[3,2,5,0,-5]
Advanced_System_Care=[1,5,0,3,-2]
Backup_pro14=[1,0,2,0,0]
````

<div class="alert alert-block alert-info">
<b>Napomena:</b> Umesto lista mogli smo da koristimo i NumPy nizove, ukoliko bismo taj modul prethodno učitali.  </div>

U Jupyter radnim sveskama sve informacije koje želimo da prikažemo na jednom grafiku, naredbe za izvršenje njihovih prikaza moraju da budu deo jedne iste kodne ćelije. Postavljanjem i brisanjem oznaka za komentare `#` u kodnoj ćeliji ovog primera možete lakše da uočite šta svaka od naredbi ponaosob dodaje grafiku.

Svaka od osa podeljena je manjim i većim oznakama [ticks](https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html) (crticama) na duži jednakih dužina. Dužina crtica ukazuje da Matplotlib podržava dva nivoa podele osa - glavni (`major`, uobičajeno aktivan) i sporedni (`minor`, uobičajeno neaktivan). Aktivni podeoci koriste se za povlačenje mreže linija `grid`. 

````python
slika, ax = plt.subplots(figsize=(8,6))          #definisanje imena objekta figure, imena odeljaka (axes) i njihovog broja-0
plt.title("Linijski prikaz prodaje i nabavke")
plt.plot(Digital_Lab, label='Digital Lab')         #opcija label imenuje podatke za legendu   
plt.xlabel("Dani u nedelji")
plt.ylabel("Prodaja po komadu")
plt.plot(Mobile_Transfer, label='Mobile Transfer')
plt.plot(Advanced_System_Care,label='Advanced System Care')
plt.plot(Backup_pro14,label='Backup Pro 14')
plt.xticks(Dani,["Ponedeljak","Utorak","Sreda","Četvrtak","Petak"])
plt.legend(loc='lower left', shadow=True)
ax.yaxis.set_minor_locator(plt.MultipleLocator(1))
plt.grid(b=True, which='major', linestyle='-')
plt.grid(b=True, which='minor', linestyle='--')
````

**Primer 3.** Neka je data funkcija $f(x)=x^2\cos(x-2)$ na domenu $x\in[-3,6].$ Nacrtaćemo linijski grafik ove funkcije na osnovu $100$ tačaka vrednosti. Zbog velikog broja vrednosti, tačke sa $x-$ose ne unosimo ručno, već geberišemo nekom naredbom za automattsko generisanje nizova (ili lista).

````python
import numpy as np
````

````python
domen=np.linspace(-3,6,100)   #kreiranje x-podatka
fx=np.cos(domen-2)*(domen**2) #kreiranje y-podatka
````

Ovim smo pripremili podatke za crtanje grafika. Ukoliko primenimo naredbu **plot** bez navođenja $x-$vrednosti dobićemo grafik koji nije u potpunosti odgovarajući.

````python
plt.plot(fx)
plt.title("Grafik van domena funkcije");
````

Ukoliko uvrstimo i tačke domena u poziv naredbe dobićemo adekvatan raspored tačaka sa grafika.

````python
plt.plot(domen,fx)
plt.title("Grafik na domenu funkcije")
plt.grid()
````

Primećujemo da je 100 tačaka podataka bilo dovoljno za generisanje grafika glatkog izgleda.

**Primer 4.** Na primeru grafika funkcije 
$$ f(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}},\qquad x\in[-5,5]$$
pokazaćemo uticaj finoće gustine uzorkovanja kontinualne funkcije na izgled linijskog grafika.

````python
from ipywidgets import interactive

def f(n):
    x = np.linspace(-5, 5, n)
    y = np.exp(-x**2/2)/np.sqrt(2*np.pi)
    plt.title("Gausovo zvono")
    plt.plot(x, y,'-')
    plt.ylim(-0.3, 0.6)
    plt.grid()

zvono = interactive(f, n=(5,50, 5))
zvono
````

**Primer 5.** Iskoristićemo specijalizovane funkcije NumPy modula za rad sa polinomima za njihove grafičke prikaze. Za početak posmatraćemo polinom $P_5(x)=3x^5-2.1x^4+1.12x^3-3.74,$ zadat svojim vektorom koeficijenata. Potražićemo njegov grafik na segmentu $[-1,2].$

````python
P5=np.array([3,-2.1,1.12,0,0,-3.74])
xk=np.linspace(-1,2)
fk=np.polyval(P5,xk)
````

**Primer 7.** U narednom primeru kombinovaćemo prethodno opisane prikaze i opcije. Primenjena je i biblioteka `random` kako bi se lakše generisali podaci za prikaz.

````python
import numpy.random as rndm
````

````python
x = np.linspace(-10., 10., 200)
y = np.sin(x) * np.exp(-(x/5.0)**2)
yerror=rndm.uniform(-0.1,0.1,[20])

xdata= np.linspace(-10., 10., 20)
ydata = np.sin(xdata) * np.exp(-(xdata/5.0)**2)+yerror

plt.figure(1, figsize=(6, 4))
plt.plot(x, y, 'g-', label='teorija') 
plt.plot(xdata, ydata, 'bo', label="podaci") 
plt.xlabel('x')
plt.ylabel('transverzalno izmeštanje')
plt.legend(loc='upper right', title='legenda')
plt.axhline(color='gray', zorder=-1)
plt.axvline(color='gray', zorder=-1);
````

Podaci prikazani na prethodnoj slici imitiraju situaciju eksperimentalnih podataka sa pokušajem njihovog modeliranja, tj. pronalaženja funkcije koja će ih pratiti na neki blizak način. Takve funkcije koristimo za približnu dopunu informacija o praćenom fenomenu. Na gornjoj slici tačkasto su prikazani mereni podaci - eksperimentalni podaci, dok je zelenom linijom dat model koji dalje vrednosti merene veličine imitira i predviđa. U takvim situacijama veoma često se praktikuje uključivanje podataka o odstupanjima vrednosti eksperimenta i njegovog modela. To nam pomaže, između ostalog, u proceni valjanosti dobijenog modela. Modul Matplotlib poseduje funkciju `errorbar` koja tačkaste podatke prikazuje sa horizontalnim i/ili vertikalnim linijama koje služe za opis greške. Narednim primerom prikazana je upotreba ove funkcije.

````python
plt.figure(1, figsize=(6, 4))
plt.plot(x, y, '-C0', label="teorija")
plt.errorbar(xdata, ydata, fmt='oC1', label="podaci",
xerr=0.75, yerr=yerror, ecolor='black')
plt.xlabel('x')
plt.ylabel('transverzalno izmeštanje')
plt.legend(loc='upper right');
````

**Primer 8.** Prikazaćemo grafik funkcije $\tan x$ za vrednosti argumenta $x\in[0,10],$ jasno je radi se o uglovima u radijanima. To možemo da pokušamo da obavimo bez mnogo uređivanja na već prikazan način, jednim pozivom naredbe `plot` bez dodatnih argumenata, kao što je urađeno u narednoj kodnoj ćeliji.

````python
x = np.arange(0.01, 10., 0.04)
ytan = np.tan(x)

plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytan);
````

Rezultujući grafik ne izgleda baš kako smo očekivali. Problem nastaje zbog toga što funkcija $\tan x$ poseduje vertikalne asimptote u tačkama $x\in\{\pm\pi/2,\pm3\pi/2,\dots\}.$ Drugim rečima veliki rasponi vrednosti se registruju i prikazuju kao šiljci na grafiku. Takođe neujednačenost grafika ukazuje na prekoračenje u predstavivim vrednostima i nestabilnost u izračunavanju. Naravno,takav grafik je od slabe koristi. Rešenje nalazimo u ograničenju skupa vrednosti $y$ koordinata koje se prikazuju na grafiku. Tome služi Matplotlib funkcija `ylim` kao što je demonstrirano u narednoj ćeliji.

````python
plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytan)
plt.ylim(-8, 8)
plt.axhline(color="gray", zorder=-1);
````

Sa gornjeg grafika možemo i da isključimo vertikalne linije koje se pojavljuju u blizini tačaka $x\in\{\pi/2,3\pi/2,5\pi/2\}.$ Tome služi naredba za [maskiranje](https://numpy.org/doc/stable/reference/generated/numpy.ma.masked_where.html) NumPy nizova. Primer upotrebe dat je u narednoj ćeliji. Prvi argument opisuje uslov za maskiranje elemenata nekog niza, sam niz se navodi kao drugi argument.

````python
ytanM = np.ma.masked_where(np.abs(ytan) > 20., ytan)

plt.figure(figsize=(8.5, 4.2))
plt.plot(x, ytanM)
plt.ylim(-8, 8) 
plt.axhline(color="gray", zorder=-1);
````

**Primer 10.** Kreiraćemo nepravilnu tabelu grafičkih elemenata. Nepravilnost se ogleda u nejednakom broju slika unutar vrsta tabela. Najpre kreiramo tačkaste podatke za grafički prikaz. Funkcije koje će biti prikazane su $\quad s(x)=\begin{cases}1,&x=0,\\ \dfrac{\sin x}{x},&x\neq0,\end{cases}\quad$ i $\quad exp(x)=e^x.$

````python
#definisanje funkcije s(x)
def s(x):
    a = np.where(x == 0., 1., np.sin(x)/x)
    return a

#podaci za funkciju s
xs = np.linspace(-15., 15., 150)
ys = s(xs)

#podaci za eksponencijalnu funkciju
xe = np.arange(0., 10., 0.1)
ye = np.exp(xe)
````

````python
# kreiranje grafičkog prozora izabrane veličine
slika = plt.figure(figsize=(9, 7))
plt.suptitle("Kombinovanje grafika 2+1")
slika.subplots_adjust(hspace=0.3, wspace = 0.3) #razmak između delova prozora

# subplot: klasičan prikaz eksponencijalne funkcije
plt.subplot(2, 2, 1)
plt.plot(xe, ye)
plt.title(r'$e^x$')

# subplot: y-log prikaz eksponencijalne funkcije
plt.subplot(2, 2, 2)
plt.plot(xe, ye)
plt.yscale("log")
plt.title(r'$e^x$ sa $\log$ razmerom na $y$ osi')

# subplot: prikaz s(x) funkcije
plt.subplot(2, 1, 2)
plt.plot(xs, ys)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.title(r's(x)=(\sin x)/x');

#slika.tight_layout()  #način razmaka između podslika
````

## Histogram

Za narednu temu biće koristan jedan poseban način predstavljanja numeričkih podataka - histogram. On predstavlja grafičku verziju tabele u kojoj se čuvaju informacije o frekvenciji pojavljivanja vrednosti iz unapred izabranih intervala. Histogram je način prikazivanja podataka raspoređenih u određene kategorije (klase) ili grupe. Pojasnićemo ovo na primeru.

**Primer 11.** Pretpostavimo da je dat vektor od 50 celobrojnih vrednosti u opsegu $[0,20].$ Želimo da saznamo frekvenciju pojavljivanja vrednosti iz pojedinačnih intervala $[0,5),[5,10),[10,15),[15,20].$ Takvom analizom možemo da utvrdimo da li su neki opsezi vrednosti zastupljeniji ili ne. 

Horizontalnim linijama na grafiku prikazana je podela vrednosti po grupama. 

````python
primer9, ax = plt.subplots()  
indeksi=np.arange(50)
vrednosti=np.random.randint(21,size=(50,))
plt.scatter(indeksi,vrednosti)
ax.yaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='y', linestyle='-')
````

Prebrojavanjem elemenata u svakoj od grupa dobijamo podatke o zastupljenosti vrednosti iz svake od grupa. Grafički se to pokazuje visinom stubaca iznad odgovarajućeg intervala.

````python
histogram, ax = plt.subplots()  
plt.hist(vrednosti, 4)
plt.title("Histogram")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='both', linestyle='-')
````

Ukoliko želimo da histagram pokazuje relativne vrednosti ponavljanja potrebno je postaviti opciju **density=True**.

````python
normalizovan_histogram, ax = plt.subplots()  
plt.hist(vrednosti, 4, density=True)
plt.title("Histogram")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(b=True, which='major', axis='both', linestyle='-')
````
