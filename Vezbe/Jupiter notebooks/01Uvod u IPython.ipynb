## Markdown ћелије


Бавићемо се најпре Markdown ћелијама. Markdown ћелије користе се као помоћ за детаљан опис креираног кода или добијених резултата, уз могућност постављања корисних линкова или других html објеката. Ћелија овог типа добија се отварањем новог простора за ћелију и дефинисањем њеног типа за Markdown. То може да се обави кроз траку алата, избором Markdown као што је приказано на претходној слици. Други начин јесте да након избора постојеће ћелије на тастатури притиснете слово "m".

Markdown представља врсту [Markup језика](https://en.wikipedia.org/wiki/Markup_language) дизајнираног за једноставно читање и писање. Користићемо их за испис текста коментара и поставке задатака. Класичан текст се најједноставније уноси, и унос је у таквој ситуацији дослован. Уколико желимо да текст добије уређенију структуру од обичних коментара, опције описане у документу 02Uredjivanje Teksta у томе помажу. Овде ћемо се задржати искључиво на опцијама које су потребне за завршавање домаћих задатака у оквиру предмета.

**Напомена:** Двоструким кликом на  ћелију моћи ћете да упознате начин како је текст у њима креиран. Због тога Markdown ћелије овог декумента представљају упутство кроз примере за неке од могућности Markup језика.

## Унос математичких симбола и формула 

Унос математичких формула и симбола користи [LaTex](https://en.wikipedia.org/wiki/LaTeX) ознаке. Да би се активирао интерпретатор за разумевање LaTex језика одговарајући делови текста морају да отпочну резервисаним симболима или наредбама. Тако нпр. једначине унутар једног реда текста почињу и завршавају се симболом $\$.$ 

Улаз:
    
    Текст са формулом у истом реду: $x^3-\cos(x).$ Наставак текста је такође у истом реду.

Излаз:
> Текст са формулом у истом реду: $x^3-\cos(x).$ Наставак текста је такође у истом реду.
    

За центриране једначине издвојене у посебан ред користе се парови симбола \$\$ на почетку и на крају уноса математичког текста.
Улаз:

        $$\frac{\sqrt{x}}{\log(x+1)}.$$

Излаз:
$$\frac{\sqrt{x}}{\log(x+1)}.$$

Када хоћемо да опишемо део-по-део дефинисане функције можемо да користимо наредну конструкцију. Она је погодна и за системе једначина.

Улаз:

    $$ \begin{align} &\text{Функција густине вероватноће:}&& \text{Сплајн функција:}&&\text{Систем једначина:}\\
    &\begin{cases} \frac{1}{b-a},&x\in[a,b]\\ 0,&\text{иначе} \end{cases}

    &&\begin{cases} 2x+3y&=5,\\ -4x+6y&=1,\\ z&=-1 \end{cases}
    \end{align}$$

Излаз:
$$
\begin{align}
&\text{Функција густине вероватноће:}&&
\text{Сплајн функција:}&&\text{Систем једначина:}\\

\end{cases}
&&
\begin{cases}
P_1(x),&\ x<a\\
P_2(x),&x\in[a,b)\\

\end{cases}
&&\begin{cases}

\end{align}
\quad
$$

**Пример 1.** Направићемо програм који користи математички модел за предвиђање висине лопте бачене вертикално према горе са површине земље. Једначина модела гласи:
$$ y=v_0t-\frac{gt^2}{2}$$


- Вертикална позиција лопте у односу на тло носи ознаку $y,$ 
- почетна брзина приликом избачаја је $v_0,$ 
- убрзање гравитације Земље је $g$
- време проведено у кретању је $t$

```python
v0=21
g=9.81
t=2
y=v0*t-0.5*g*t**2
print(y)
```

Уколико желимо да експериментишемо са различитим вредностима параметара овог модела, нпр. различите почетне брзине или различити тренуци у којима проверавамо положај лопте, претходни програм би требало више пута активирати са различитим улазним параметрима. Уместо да се вредности променљивих $v_0$ и $t$ мењају ручним уносом, интерактивна промена вредности добија се употребом тзв. widget-a. Пример претходног кода са активацијом widget-а дат је у наставку. Кодна ћелија мора да се активира како би излаз са клизачима био активан за употребу.

```python
import ipywidgets as widgets

#v0=18
g=9.81
#t=3

def y(v0,t):
    return v0*t-0.5*g*t**2

widgets.interact(y, v0=widgets.IntSlider(min=10, max=30, step=1, value=18), 
                 t=widgets.IntSlider(min=1.0, max=5.0, step=1, value=2.0))
```

**Пример 2.** Померањем клизача у пољу за  widget могуће је да се добије и интерактивни график, не само промена вредности једне променљиве. Овакви примери су веома корисни у нумеричким и графичким експериментима. Наредним кодом дат је пример једног интерактивног графика праве. Мењањем вредности коефицијената можемо да утичемо на карактеристике графика.

```python
from ipywidgets import interactive
import matplotlib.pyplot as slika
import numpy as np

def f(k, n):
    print("Пример интерактивног графика")
    slika.figure(2)
    x = np.linspace(-10, 10, num=1000)
    slika.plot(x, k * x + n)
    slika.ylim(-5, 5)

interactive_plot = interactive(f, k=(-3.0, 2.0), n=(-3, 3, 0.5))
interactive_plot
```
