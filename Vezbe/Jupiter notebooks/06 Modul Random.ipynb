### rand  i uniform

Naredba [rand](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html) ima sintaksni oblik
```python
numpy.random.rand()
```
ili
```python
numpy.random.rand(dim1,dim2,...)
```
Prvi nain pozivanja slui za generisanje jednog sluajnog broja iz opsega $[0,1).$

````python
rndm.rand()
````

Drugi nain pozivanja naredbe generie n-dimenzionalni niz sluajnih brojeva iz opsega $[0,1).$ Ulazne vrednosti dim1,dim2,... odreuju dimenzije ovog niza.

````python
rndm.rand(2,3)
````

````python
rndm.rand(6,)
````

````python
rndm.rand(6)
````

Brojevi koji se generiu narredbom `rand` podleu tzv. uniformnoj raspodeli. To znai da bi histogram vrednosti ovako generisanog vektora (za velike dimenzije) imao mala odstupanja u visinama pravougaonika histograma. 

````python
vrednosti=rndm.rand(10000,)
plt.hist(vrednosti, 10);
d=1-0;              #duina segmenta sa koga se biraju vrednosti
````

````python
plt.hist(vrednosti, 10,density=True)
plt.plot([1/d,1/d],'--r');
````

Ukoliko je potrebno da se generie n-dimenzionalni niz sa istim karakteristikama, ali na drugom opsegu vrednosti, to se postie naredbom [uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html).

````python
min, max=1, 101
vrednosti=rndm.uniform(min,max,10000)
plt.hist(vrednosti, 10, facecolor='green');
d=max-min;              #duina segmenta sa koga se biraju vrednosti
````

````python
plt.hist(vrednosti, 10, facecolor='green',density=True)
plt.plot([min,max],[1/d,1/d],'--r');
````

````python
rndm.uniform(-1,1,[2,3])
````

### randint

Naredbom [randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) se na slian nain generiu vektori celobrojnih vrednosti sa diskretnom uniformnom raspodelom. Moe se koristiti za dobijanje jednog sluajnog broja ili skupa brojeva.

````python
rndm.randint(5,21)                #generisanje jedne celobrojne sluajne vrednosti u opsegu [5,20]
````

````python
min, max=10,46
vrednosti=rndm.randint(min, max,(1000,))
plt.hist(vrednosti, 6);
````

````python
d=max-min;
plt.hist(vrednosti, 6, density=True)
plt.plot([min,max],[1/d,1/d],'--r');
````

### randn i normal

Sledee dve naredbe kreiraju jednu sluajnu vrednost ili niz sluajnih vrednosti koje imaju tendenciju nagomilavanja u izabranom opsegu. Odgovarajui histogrami podseaju na grafik Gausovog zvona
$$ f(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2},\qquad x\in\mathbb{R}$$
Vrednosti se grupiu oko broja $\mu,$ i najvei broj vrednosti nalazi se u opsegu $[\mu-\sigma,\mu+\sigma].$ Za takve brojeve kaemo da podleu normalnoj raspodeli.

Naredba [randn](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) generie sluajne brojeve sa normalnom raspodelom i parametrima $\mu=0$ i $\sigma=1.$ 

Naredbom [normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) mogue je generisati sluajne brojeve sa normalnom raspodelom i proizvoljnim vrednostima parametara $\mu$ i $\sigma.$

````python
vrednosti=rndm.randn(12345)
````

````python
vrednosti.mean()
````

````python
vrednosti.std()
````

````python
plt.hist(vrednosti, 30, density=True);
x = np.linspace(-3, 3, 70)
y = np.exp(-x**2/2)/np.sqrt(2*np.pi)
plt.plot(x, y,'-r');
````

````python
plt.hist(vrednosti, 30, density=True);
x = np.linspace(-3, 3, 70)
y = np.exp(-x**2/2)/np.sqrt(2*np.pi)
plt.plot(x, y,'-r');
````

````python
mu, sigma=5,6
vrednosti=rndm.normal(mu,sigma,12345)
plt.hist(vrednosti, 10, density=True);
x = np.linspace(mu-4*sigma, mu+4*sigma, 70)
y = np.exp(-((x-mu)/sigma)**2/2)/(sigma*np.sqrt(2*np.pi))
plt.plot(x, y,'-r');
````

````python
plt.hist(vrednosti, 50, density=True);
x = np.linspace(mu-4*sigma, mu+4*sigma, 70)
y = np.exp(-((x-mu)/sigma)**2/2)/(sigma*np.sqrt(2*np.pi))
plt.plot(x, y,'-r');
````

````python
np.mean(vrednosti)
````

````python
np.std(vrednosti)
````

**Primer 1.** U prethodnoj radnoj svesci posmatrali smo uticaj broja uzoraka funkcije na glatkost njenog grafika. Sada emo da razmotrimo uticaj greaka na izgled uzorka. Skiciraemo grafik funkcije 
$$f(x)=x^4-4\cos(x),\qquad x\in[0,1]$$
Uzorkovaemo funkciju na malom broju vrednosti domena, a zatim emo da ih "pokvarimo" ubacivanjem namerne-sluajne greke.

````python
from ipywidgets import interactive

greska=rndm.uniform(-0.5,0.5,10)
````

````python
def f(r):
    x = np.linspace(0, 1, 70)
    y = x**4-4*np.cos(x)
    plt.plot(x, y)
    xp=np.linspace(0,1,10)
    yp = xp**4-4*np.cos(xp)+r*greska
    plt.plot(xp, yp,'o')

fuja = interactive(f, r=(0.1,1, 0.05))
fuja
````

U praksi greka se najee modeluje Gausovom, tj. normalnom raspodelom. Razmotrimo i ovaj sluaj grafiki.

````python
greska=rndm.normal(0,0.5,10)
````

````python
fuja = interactive(f, r=(0.1,1, 0.05))
fuja
````
